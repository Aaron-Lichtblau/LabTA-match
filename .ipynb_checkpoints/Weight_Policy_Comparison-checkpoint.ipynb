## **Definitions of Stats being used:**

students = i

$shift(j, i)$ = shifts given in schedule to student i

$pref(i, j)$ = student i's preference score of slot j

$cap(i)$ = student i's cap on number of shifts they want to work

**happiness$_i$ =** $100 * \frac{\sum_{shift(j, i)}{pref(i, j)}}{3 * cap(i)}$

   Intuition: the most happy a student can be is if they get the number of shifts they want (cap) and all of the shifts they get are preferenced as 3's. Average happiness will be reported as the average over all these student happiness values.

**variance of happiness:** standard definition of variance, where happiness scores range from [0, 1].

**availability-happiness correlation:** availability$_i$ is a summation of the preference scores that student i gave. This statistic is the correlation value between these availability scores and happiness scores of students.

**min/most happy people:** report of the students whose happiness scores were outliers (following the 3 z-score rule)

## **Definitions of Weights being used:**

slot_type: controls how much students' requested slot type (2hr vs 4hr or 0 = no preference) is taken into account by the algorithm

no_1: controls how giving out 1's is regulated

guarantee_shift: controls guarantee that every student gets a shift

avail: controls how students' availability scores get rewarded

shift cap: controls how much students' caps on shifts get rewarded

equality: controls how equal the number of shifts given out to students is (opposite of shift cap)

## **Note on Schedule Inputs**

These inputs are the defaults based on the Fall 2019 LabTA Schedule. Please edit the dictionaries as needed for changes in the schedule itself.

```python
min_gap = 3 #minimum amount of time between shifts (end time to start time) for them to not overlap
duration = 2 #length of slots (in hours)

#dict of target number of students in each slot
slotdict = {"M_7" : 8, "M_9" : 6,"Tu_7" : 4, "Tu_9" : 4,"W_7" : 4, "W_9" : 4,"Th_7" : 5, "Th_9" : 4,"F_7" : 4, "F_9" : 4,"Sa_3" : 5, "Sa_4" : 6,"Sa_5" : 5,"Su_5" : 4,"Su_6" : 3,"Su_7" : 6,"Su_8" : 4, "Su_9" : 6}

#list of slots that need more skilled TA's
stress_slots = ["M_7", "M_9", "Su_8", "Su_9"]

#numeric value indicating how many TAs the scheduler can hire above the targeted value for any given slot
target_delta = 2

#number of shifts the scheduler can assign in addition to the slotdict shift numbers
flex_shifts = 4

#sets minimum number of experienced TA's per slot
MIN_EXP = 3

#sets minimum number of skilled TA's per stress slot
MIN_SKILL = 2

#gets number of slots
NUM_SLOTS = 0
for slot in slotdict:
    NUM_SLOTS += slotdict[slot]
```

```python
def color_schedule(val):
    if int(val) == -3.0:
        background_color = 'green'
    elif int(val) == -2.0:
        background_color = 'yellow'
    elif int(val) == -1.0:
        background_color = 'red'
    else:
        background_color = ''
    return 'background-color: %s' % background_color
```

```python
def color_wrong_type(s):
    '''
    highlight the wrong types dark orange
    '''
    is_wrong = s == True
    return ['background-color: darkorange' if v else '' for v in is_wrong]
```

```python
def results(weight_dict):
    df = input_creator.get_df('LabTA_test2 - Sheet1.csv')

    #dict of slots to check as keys, and overlapping slots as values (student won't be placed in overlap)
    slots = input_creator.get_slots(df)
    OVERLAPS = input_creator.get_overlaps(slots, min_gap, duration)
    #dict of slots and their prev slots
    PREV_SLOT = input_creator.get_prev_slots(df, duration)
    print(PREV_SLOT)
    
    #print missing slots from slotdict (to guarantee 2 slots per person)
    missing_slots = max_weight_cap.check_slotdict(df, slotdict, flex_shifts)
    print('There are ', missing_slots, " fewer slots in slotdict than needed to guarantee 2 slots for every TA")

    #get experience dict and fake temp skill dict for testing
    exp_dict = {}
    skill_dict = {}
    students = list(df['name'])
    for index in range(len(df['name'])):
        exp_dict[str(df.at[index, 'name'])] = int(df.at[index, 'experience'])
        skill_dict[str(df.at[index, 'name'])] = int(df.at[index, 'skill'])
    
    #create graph nodes and weight edges
    graph_data = max_weight_cap.create_graph(df, weight_dict, slotdict, PREV_SLOT, NUM_SLOTS)
    student_nodes = graph_data[0]
    slot_nodes = graph_data[1]
    wt = graph_data[2]
    
    #solve the problem and get the ordered schedule
    p = max_weight_cap.solve_wbm(student_nodes, slot_nodes, wt, slotdict, PREV_SLOT, OVERLAPS, MIN_EXP, exp_dict, MIN_SKILL, skill_dict, stress_slots, target_delta, flex_shifts)
    unordered_sched_dict = max_weight_cap.get_solution(p)
    max_weight_sched = max_weight_cap.order_sched(df, unordered_sched_dict, slotdict)

    #print df, exp and happiness stats
    max_weight_cap.schedule_to_df(df, max_weight_sched)    

    post_hap = stats.sched_happiness(df, max_weight_sched, PREV_SLOT)
    df['wrong type'] = post_hap[8]
    cols = ['name', 'M_7', 'M_9', 'Tu_7', 'Tu_9', 'W_7', 'W_9', 'Th_7', 'Th_9', 'F_7', 'F_9', 'Sa_3', 'Sa_4', 'Sa_5', 'Su_5', 'Su_6', 'Su_7', 'Su_8', 'Su_9', 'happiness', 'wrong type', 'slot_type', 'availability', 'cap', 'experience', 'skill', 'hours']
    df = df[cols]
    styled_df = df.style.\
        applymap(color_schedule, subset = pd.IndexSlice[:, max_weight_sched.schedule.keys()]).\
        apply(color_wrong_type, subset=['wrong type'])
    
    display(styled_df)
    
    print("Max Weight schedule stats:")
    stats.exp_stats(exp_dict, max_weight_sched)
    stats.skill_stats(skill_dict, max_weight_sched)
    
    #Evaluate happiness stats of schedule
    print('Average Happiness: ', post_hap[0])
    print()
    print('Availability to happiness correlation: ', post_hap[1])
    print()
    print('Variance of happiness: ', post_hap[2])
    print()
    print('Min happy outlier students: ', post_hap[3])
    print()
    print('Max happy outlier students: ', post_hap[4])
    print()
    print('Students who got 1\'s: ', post_hap[5])
    print()
    print('Students without shifts: ', post_hap[6])
    print()
    print('Students who got wrong slot type: ', post_hap[7])
    print()
    
```

## **Default Policy**

### This policy guarantees that no 1's get used, everyone gets a shift, and puts strong emphasis on rewarding students who give high availability

```python
#Default weights
weight_dict = {}
weight_dict['slot_type'] = 4
weight_dict['no_1'] = 9
weight_dict['guarantee_shift'] = 8
weight_dict['avail'] = 5
weight_dict['shift_cap'] = 3
weight_dict['equality'] = 6
```

```python
results(weight_dict)
```

## **Slot Type Policy**

### This policy guarantees that all students get a shift, maximizes giving students the slot type they requested, then minimizes 1's given, and puts some emphasis on rewarding students who give high availability

```python
#Slot type weights
weight_dict_slot = {}
weight_dict_slot['slot_type'] = 7
weight_dict_slot['no_1'] = 6
weight_dict_slot['guarantee_shift'] = 8
weight_dict_slot['avail'] = 4
weight_dict_slot['shift_cap'] = 2
weight_dict_slot['equality'] = 3
```

```python
results(weight_dict_slot)
```

## **Power Student Policy**

### This policy guarantees that all students get a shift, maximizes giving shifts to students who have higher caps and availabilities, and puts some emphasis on giving students the slot type that they requested.

```python
#power student weights
weight_dict_power = {}
weight_dict_power['slot_type'] = 4
weight_dict_power['no_1'] = 3
weight_dict_power['guarantee_shift'] = 8
weight_dict_power['avail'] = 6
weight_dict_power['shift_cap'] = 7
weight_dict_power['equality'] = 3
```

```python
results(weight_dict_power)
```

## **Note on how to set weights:**

Put an int from 1-9 for each category of weight. Values given >5 will be treated as strict constraints: they will be adhered to strictly in the order that they are valued. For example, if slot_type = 9, all students will be given slots that are of their requested slot type (2hr vs 4hr). Values <=5 will be treated as soft constraints. For example, if avail = 5, students with higher availabilities will be given an advantage in the schedule, however this advantage only comes into effect after all strict constraints are met, and the advantage can be outweighed by other factors (sums of other soft constraints).

```python
#custom weights
weight_dict_custom = {}
weight_dict_custom['slot_type'] = 5
weight_dict_custom['no_1'] = 7
weight_dict_custom['guarantee_shift'] = 6
weight_dict_custom['avail'] = 3
weight_dict_custom['shift_cap'] = 5
weight_dict_custom['equality'] = 3
```
